## ADDITIONAL FEATURES
Additional features of the `verify_barcode` function.

### Visualization

```python
detection_dict, rotation_dict, refinement_dict, overall_quality_parameters_dict = verify_barcode.verify_barcode(image_path,
                                                                                                 visualization_dict='all')
```

We can also specify to visualize only some specific plots.

Example: visualize only the refined ROI image with the $10$ scanlines and the scanlines reflectance profiles.

```python
visualization_dict = {
    'visualize_scanlines_onRoiImage': True,
    'visualize_scanlines_qualityParameters': True
}
detection_dict, rotation_dict, refinement_dict, overall_quality_parameters_dict = verify_barcode.verify_barcode(image_path,
                                                                                    visualization_dict=visualization_dict)
```

### Timing information

```python
detection_dict, rotation_dict, refinement_dict, overall_quality_parameters_dict = verify_barcode.verify_barcode(image_path,
                                                                                                 verbose_timing=True)
```

### Building the output file

```python
detection_dict, rotation_dict, refinement_dict, overall_quality_parameters_dict = verify_barcode.verify_barcode(image_path,
                                                                                               create_output_file=True,
                                                                                               output_folder_path='../out')
```

### Other parameters

```python
detection_dict, rotation_dict, refinement_dict, overall_quality_parameters_dict = verify_barcode.verify_barcode(image_path,
                                                                                    n_scanlines=20,
                                                                                    compute_barcode_structure_algorithm=4)
```
